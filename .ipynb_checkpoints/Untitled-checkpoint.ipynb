```python
import csv
import matplotlib.pyplot as plt

years = []
alcohol_consumption = []
students = []

with open('istherecorrelation.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=';')
    next(csv_file)
    for row in csv_reader:
        years.append(row[0])
        students.append(row[1])
        alcohol_consumption.append(row[2])
        
plt.plot(years, alcohol_consumption)
plt.plot(years, students)
plt.savefig("correlation.png")
plt.show()
```
